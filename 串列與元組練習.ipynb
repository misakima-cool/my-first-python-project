```python
# 🎒 魔法背包：串列基礎操作示範

print("🎒 魔法背包：串列基礎操作")
print("=" * 30)

# 1️⃣ 創建串列（建立魔法背包）
# print()
print("\n1️⃣ 創建魔法背包：")

# 空背包
empty_bag = []
print(f"   空背包：{empty_bag}")

# 裝了東西的背包
school_bag = ["課本", "筆記本", "鉛筆", "橡皮擦"]
print(f"   書包：{school_bag}")

# 數字背包
lucky_numbers = [7, 3, 11, 21, 42]
print(f"   幸運數字：{lucky_numbers}")

# 混合背包（串列可以裝不同類型的物品！）
mixed_bag = ["蘋果", 3, "香蕉", 5.5, True]
print(f"   混合包：{mixed_bag}")

# 2️⃣ 查看背包內容
print("\n2️⃣ 查看背包內容：")
fruits = ["🍎蘋果", "🍌香蕉", "🍇葡萄", "🍓草莓", "🥝奇異果"]

print(f"   水果籃：{fruits}")
print(f"   第一個水果：{fruits[0]}")
print(f"   最後一個水果：{fruits[-1]}")
print(f"   中間三個：{fruits[1:4]}")
print(f"   背包大小：{len(fruits)} 個物品")

# 3️⃣ 加入新物品
print("\n3️⃣ 加入新物品：")

# 在最後加入
fruits.append("🥭芒果")
print(f"   加入芒果後：{fruits}")

# 在指定位置插入
fruits.insert(2, "🍊橘子")
print(f"   在第3位插入橘子：{fruits}")

# 一次加入多個
more_fruits = ["🍑櫻桃", "🍈哈密瓜"]
fruits.extend(more_fruits)  #擴充
print(f"   批量加入後：{fruits}")

# 4️⃣ 移除物品
print("\n4️⃣ 移除物品：")

# 移除指定物品
fruits.remove("🍊橘子")
print(f"   移除橘子後：{fruits}")

# 彈出最後一個
last_fruit = fruits.pop()
print(f"   彈出最後一個：{last_fruit}")
print(f"   剩餘水果：{fruits}")

# pop 也可以指定"位置"移除唷
index_fruit = fruits.pop(2)
print(f"   彈出指定位置：{index_fruit}")
print(f"   剩餘水果：{fruits}")


# 5️⃣ 修改物品
print("\n5️⃣ 修改物品：")
fruits[0] = "🍎紅蘋果"
print(f"   把蘋果變成紅蘋果：{fruits}")

print(f"\n🎉 最終的水果籃：{fruits}")
```

```python
# 🎪 串列進階魔法

print("🎪 串列進階魔法")
print("=" * 20)

# 1️⃣ 排序魔法
print("\n1️⃣ 排序魔法：")
numbers = [64, 34, 25, 12, 22, 11, 90]
print(f"   原始數字：{numbers}")

# 從小到大排序（要先複製，不然原本的會改變）
numbers_asc = numbers.copy()
numbers_asc.sort()
print(f"   從小到大：{numbers_asc}")

# 從大到小排序
numbers_desc = numbers.copy()
numbers_desc.sort(reverse=True)
print(f"   從大到小：{numbers_desc}")

# 2️⃣ 搜尋魔法
print("\n2️⃣ 搜尋魔法：")
students = ["小明", "小華", "小美", "小強", "小雯"]
print(f"   學生名單：{students}")

# 檢查是否存在
print(f"   小明在班上嗎？ {'小明' in students}")
print(f"   小陳在班上嗎？ {'小陳' in students}")

# 找到位置
position = students.index("小美")
print(f"   🔍 小美在第 {position} 個位置（從0開始算）")

# 計算出現次數
grades = ["A", "B", "A", "C", "A", "B", "A"]
a_count = grades.count("A")
print(f"   📊 成績統計：A等有 {a_count} 個")

# 3️⃣ 串列組合魔法
print("\n3️⃣ 組合魔法：")
list1 = ["紅", "綠"]
list2 = ["藍", "黃"]

# 串列相加
colors = list1 + list2
print(f"   顏色組合：{colors}")

# 串列重複
repeated = ["🎵"] * 5
print(f"   音符重複：{repeated}")

# 4️⃣ 串列小技巧
print("\n4️⃣ 實用小技巧：")

# 產生連續數字
numbers_1_to_10 = list(range(1, 11))
print(f"   1到10：{numbers_1_to_10}")

# 找最大值和最小值
test_scores = [85, 92, 78, 96, 88]
highest = max(test_scores)
lowest = min(test_scores)
average = sum(test_scores) / len(test_scores)
print(f"   考試成績：{test_scores}")
print(f"   最高分：{highest}，最低分：{lowest}，平均：{average:.1f}")



# max、min 也是可以運用在str 字串上唷，只是不直覺。
# 字串比大小，是比第一個字的Unicode碼
# 可查詢 Unicode https://www.unicode.org/charts/unihan.html
# 在python中可以使用ord() 指令完成

colors = ['紅', '綠', '藍', '黃']
print(f'清單：{colors}')
print(f'清單第0個位置的Unicode：{ord(colors[0])}')
print(f'清單第1個位置的Unicode：{ord(colors[1])}')
print(f'清單第2個位置的Unicode：{ord(colors[2])}')
print(f'清單第3個位置的Unicode：{ord(colors[3])}')
print(f'最大的 Unicode 是中文字：{max(colors)}')
print(f'最小的 Unicode 是中文字：{min(colors)}')

```

```python
# 💎 寶盒收藏：元組基礎操作示範

print("💎 寶盒收藏：元組基礎操作")
print("=" * 30)

# 1️⃣ 創建元組（建立寶盒）
print("\n1️⃣ 創建寶盒：")

# 空寶盒
empty_box = ()
print(f"   空寶盒：{empty_box}")

# 個人資料寶盒
person_info = ("張三", 25, "工程師", "台北")
print(f"   個人資料：{person_info}")

# 座標寶盒
coordinates = (25.0338, 121.5645)
print(f"   台北101座標：{coordinates}")

# 顏色寶盒
red_color = (255, 0, 0)
green_color = (0, 255, 0)
blue_color = (0, 0, 255)
print(f"   紅色RGB：{red_color}")
print(f"   綠色RGB：{green_color}")
print(f"   藍色RGB：{blue_color}")

# ⚠️ 特殊：單元素元組（注意逗號很重要！）
single_item = (42,)  # 這個逗號不能省略！
print(f"   單元素元組：{single_item}")

# 2️⃣ 查看寶盒內容
print("\n2️⃣ 查看寶盒內容：")

# 學生成績記錄
student_record = ("小明", "數學", 95, "2024-01-15")
print(f"   成績記錄：{student_record}")

# 使用索引取值（和串列一樣）
name = student_record[0]
subject = student_record[1]
score = student_record[2]
date = student_record[3]

print(f"   學生姓名：{name}")
print(f"   考試科目：{subject}")
print(f"   考試分數：{score}")
print(f"   考試日期：{date}")

# 3️⃣ 元組解包（超實用的技巧！）
print("\n3️⃣ 寶盒解包魔法：")

# 直接解包（一次取出所有值）
student_name, exam_subject, exam_score, exam_date = student_record
print(f"   解包結果：{student_name} 在 {exam_date} 的 {exam_subject} 考了 {exam_score} 分")

# 座標解包
latitude, longitude = coordinates
print(f"   緯度：{latitude}, 經度：{longitude}")

# RGB解包
r, g, b = red_color
print(f"   紅色分量：{r}, 綠色分量：{g}, 藍色分量：{b}")

# 4️⃣ 元組運算
print("\n4️⃣ 寶盒運算：")

# 長度
print(f"   個人資料長度：{len(person_info)}")

# 檢查是否存在
print(f"   張三在資料中嗎？ {'張三' in person_info}")

# 計算出現次數
numbers = (1, 2, 3, 2, 4, 2, 5)
count_2 = numbers.count(2)
print(f"   數字2出現了 {count_2} 次")

# 找到位置
index_3 = numbers.index(3)
print(f"   數字3在第 {index_3} 個位置（從0開始算）")

# 5️⃣ 元組組合
print("\n5️⃣ 寶盒組合：")

tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
combined = tuple1 + tuple2
print(f"   組合結果：{combined}")

# 元組重複
repeated = ("⭐",) * 5
print(f"   星星重複：{repeated}")

# 6️⃣ 元組的不可變性
print("\n6️⃣ 元組不可變特性：")
print(f"   原始個人資料：{person_info}")

# 嘗試修改元組會出錯！
print("   如果嘗試修改元組，會產生 TypeError 錯誤")
print("   例如：person_info[0] = '李四' 是不被允許的")
print("   ❌ 元組不能修改！這就是它的特色")

print("\n💡 記住：串列可以改，元組不能改！")
```

```python
# ⚖️ 串列 vs 元組大比拼

print("⚖️ 串列 vs 元組大比拼")
print("=" * 25)

# 1️⃣ 可變性比較
print("\n1️⃣ 可變性測試：")

# 串列 - 可以修改
shopping_list = ["牛奶", "麵包", "雞蛋"]
print(f"   原始購物清單：{shopping_list}")
shopping_list.append("蘋果")  # 可以新增
shopping_list[0] = "豆漿"    # 可以修改
print(f"   修改後清單：{shopping_list}")
print("   ✅ 串列可以自由修改！")

# 元組 - 不能修改
birth_info = ("王小明", "1990-05-15", "台北")
print(f"\n   出生資料：{birth_info}")
print("   如果嘗試修改元組，例如：birth_info[0] = '李小明'")
print("   ❌ 會產生 TypeError 錯誤，因為元組不能修改")

# 2️⃣ 使用場景比較
print("\n2️⃣ 使用場景對比：")

print("   📋 適合用串列 (List) 的情況：")
todo_list = ["寫報告", "買菜", "運動"]
print(f"   • 待辦事項：{todo_list} (會新增/刪除)")

student_scores = [85, 92, 78, 96]
print(f"   • 學生成績：{student_scores} (會更新分數)")

inventory = ["iPhone", "iPad", "MacBook"]
print(f"   • 庫存商品：{inventory} (會進貨/出貨)")

print("\n   💎 適合用元組 (Tuple) 的情況：")
rgb_red = (255, 0, 0)
print(f"   • RGB顏色值：{rgb_red} (固定組合)")

taipei_101 = (25.0338, 121.5645)
print(f"   • 地理座標：{taipei_101} (位置固定)")

person_basic = ("張三", "A123456789", "1985-03-20")
print(f"   • 基本資料：{person_basic} (不會改變)")

# 3️⃣ 記憶體使用比較
print("\n3️⃣ 效能比較：")

test_list = [1, 2, 3, 4, 5] * 20  # 100 個元素
test_tuple = tuple(test_list)

# 檢查記憶體大小（使用 sys 模組，但這裡簡化說明）
print(f"   💾 記憶體使用（概念說明）：")
print(f"      串列通常使用較多記憶體（因為需要支援修改功能）")
print(f"      元組使用較少記憶體（因為結構簡單固定）")
print(f"      元組存取速度通常比串列快")

# 4️⃣ 實用範例
print("\n4️⃣ 實用範例：")

# 使用元組儲存學生基本資料（不變的）
student_info = ("小王", "S001", "資工系")
name, student_id, department = student_info  # 元組解包
print(f"   學生基本資料：{name}, {student_id}, {department}")

# 使用串列儲存成績（會變的）
scores = [85, 92, 78]
print(f"   學期成績：{scores}")
scores.append(88)  # 新增期末考成績
print(f"   加入期末考：{scores}")

# 5️⃣ 選擇建議
print("\n5️⃣ 選擇建議：")
print("   🤔 選擇串列當：")
print("      • 資料會改變（增加、刪除、修改）")
print("      • 需要排序功能")
print("      • 資料量會動態變化")

print("\n   💡 選擇元組當：")
print("      • 資料固定不變")
print("      • 需要更好的效能")
print("      • 儲存設定檔或設定值")
print("      • 用來儲存座標、顏色等固定組合")

# 6️⃣ 記憶口訣
print("\n6️⃣ 記憶口訣：")
print("   📝 串列像筆記本 - 可以隨時寫入和修改")
print("   🏺 元組像古董 - 珍貴固定，不可更改")
print("   💡 資料會變用串列，固定不變用元組！")
```

```python
# 📈 範例 3：股票投資組合分析系統（簡化版）

print("📈 股票投資組合分析系統")
print("=" * 25)

# 股票基本資料：用元組儲存（股票代號、公司名稱、產業別）
stock1 = ("2330", "台積電", "半導體")
stock2 = ("2317", "鴻海", "電子製造")
stock3 = ("2454", "聯發科", "半導體")

print("\n📋 可投資股票：")
code1, name1, industry1 = stock1
print(f"   {code1} - {name1} ({industry1})")

code2, name2, industry2 = stock2
print(f"   {code2} - {name2} ({industry2})")

code3, name3, industry3 = stock3
print(f"   {code3} - {name3} ({industry3})")

# 投資組合：用串列儲存（股票代號, 持有股數, 買入價格, 當前價格）
portfolio = []

print("\n🎬 建立投資組合：")

# 新增股票到投資組合
holding1 = ("2330", 1000, 450.0, 520.0)  # 台積電
holding2 = ("2317", 2000, 85.0, 92.5)    # 鴻海
holding3 = ("2454", 500, 680.0, 725.0)   # 聯發科

portfolio.append(holding1)
portfolio.append(holding2)
portfolio.append(holding3)

print(f"   ✅ 已加入：{name1} {holding1[1]}股")
print(f"   ✅ 已加入：{name2} {holding2[1]}股")
print(f"   ✅ 已加入：{name3} {holding3[1]}股")

# 顯示投資組合詳細資訊
print("\n💼 我的投資組合：")
print("-" * 65)
print(f"{'股票':<12} {'股數':<6} {'買入價':<8} {'現價':<8} {'投資額':<10} {'市值':<10}")
print("-" * 65)

# 分析第一支股票
code, shares, buy_price, current_price = portfolio[0]
investment1 = shares * buy_price
market_value1 = shares * current_price
profit_loss1 = market_value1 - investment1

print(f"{name1}({code}) {shares:<6} {buy_price:<8.1f} {current_price:<8.1f} "
      f"{investment1:<10,.0f} {market_value1:<10,.0f}")

# 分析第二支股票
code, shares, buy_price, current_price = portfolio[1]
investment2 = shares * buy_price
market_value2 = shares * current_price
profit_loss2 = market_value2 - investment2

print(f"{name2}({code}) {shares:<6} {buy_price:<8.1f} {current_price:<8.1f} "
      f"{investment2:<10,.0f} {market_value2:<10,.0f}")

# 分析第三支股票
code, shares, buy_price, current_price = portfolio[2]
investment3 = shares * buy_price
market_value3 = shares * current_price
profit_loss3 = market_value3 - investment3

print(f"{name3}({code}) {shares:<6} {buy_price:<8.1f} {current_price:<8.1f} "
      f"{investment3:<10,.0f} {market_value3:<10,.0f}")

# 計算總計
total_investment = investment1 + investment2 + investment3
total_market_value = market_value1 + market_value2 + market_value3
total_profit_loss = total_market_value - total_investment

print("-" * 65)
print(f"{'總計':<12} {'':<6} {'':<8} {'':<8} "
      f"{total_investment:<10,.0f} {total_market_value:<10,.0f}")

print(f"\n💰 投資總結：")
print(f"   總投資金額：${total_investment:,.0f}")
print(f"   目前市值：${total_market_value:,.0f}")
print(f"   總損益：${total_profit_loss:,.0f}")

# 計算報酬率
total_return_rate = (total_profit_loss / total_investment) * 100
print(f"   總報酬率：{total_return_rate:.1f}%")

# 個股表現
print("\n🏆 個股表現：")
return1 = ((current_price - buy_price) / buy_price) * 100 if buy_price > 0 else 0
return2 = ((portfolio[1][3] - portfolio[1][2]) / portfolio[1][2]) * 100
return3 = ((portfolio[2][3] - portfolio[2][2]) / portfolio[2][2]) * 100

print(f"   {name1}: {return1:.1f}%")
print(f"   {name2}: {return2:.1f}%")
print(f"   {name3}: {return3:.1f}%")

# 找出表現最好的股票
best_return = max(return1, return2, return3)
print(f"   表現最佳：{best_return:.1f}%")

# 產業分析
print("\n🏭 產業分析：")
semiconductor_value = market_value1 + market_value3  # 台積電 + 聯發科
electronics_value = market_value2                    # 鴻海

print(f"   半導體產業：${semiconductor_value:,.0f}")
print(f"   電子製造業：${electronics_value:,.0f}")

# 計算比例
semiconductor_percentage = (semiconductor_value / total_market_value) * 100
electronics_percentage = (electronics_value / total_market_value) * 100

print(f"   半導體占比：{semiconductor_percentage:.1f}%")
print(f"   電子製造占比：{electronics_percentage:.1f}%")

print("\n✅ 股票投資組合分析完成！")
print("💡 重點：股票資訊用元組（不變），投資組合用串列（可調整）")
```
